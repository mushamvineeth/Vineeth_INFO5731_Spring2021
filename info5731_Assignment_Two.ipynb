# **Question 2**

(30 points). Write a python program to **clean the text data** you collected above and save the data in a new column in the csv file. The data cleaning steps include:

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the [stopwords list](https://gist.github.com/sebleier/554280).

(4) Lowercase all texts

(5) Stemming. 

(6) Lemmatization.

```python
# Write your code here
import nltk
nltk.download('punkt')
from nltk.corpus import stopwords
nltk.download('stopwords')

#Removing noise
covid_tweets['after removing the  punctuations']=covid_tweets['tweets'].apply(lambda x: " ".join(x for x in x.split()))


#Removing numbers
covid_tweets['after removing the numbers']=covid_tweets['tweets'].str.replace('[0-9]','')


#Removing stopwords
stop = stopwords.words('english')
covid_tweets['after removing the stopwords']=covid_tweets['tweets'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))


#Lowercase all text
covid_tweets['after lower casing all text']=covid_tweets['tweets'].apply(lambda x: " ".join(x.lower() for x in x.split()))


#Stemming
from nltk.stem import PorterStemmer
st = PorterStemmer()
covid_tweets['after stemming']=covid_tweets['tweets'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))



#Lemmatization
from textblob import Word
nltk.download('wordnet')
covid_tweets['after Lemmatization']=covid_tweets['tweets'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))



covid_tweets

```

```python
# Write your code here

#POS Tagging

nltk.download('averaged_perceptron_tagger')
nltk.download('universal_tagset')
parts_of_speech=[]
count=[]
k=0
for j in covid_tweets['tweets']:
  token=nltk.word_tokenize(j)
  parts_of_speech.append(nltk.pos_tag(token,tagset='universal'))
  count.append(nltk.FreqDist(tag for (word, tag) in parts_of_speech[0]))
  print(parts_of_speech[k])
  print(count[k])
  k=k+1
print(count)


```
